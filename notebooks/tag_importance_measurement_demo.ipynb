```python
#dependencies
import sys
sys.path.append('..')
import csv
import json
import pandas as pd
from src import constants
import matplotlib.pyplot as plt
from collections import Counter, defaultdict

#directory / file paths
data_directory = constants.DATA_DIR
clean_directory = constants.CLEAN_DIR
thumbnail_directory = constants.THUMBNAIL_DIR
ot_file_suffix = constants.img_ot_filename
st_file_suffix = constants.img_st_filename
demo_ot_file_path = f'{data_directory}/{ot_file_suffix}_13001.json'
demo_st_file_path = f'{data_directory}/{st_file_suffix}_13001.json'
img_summary_path = f'{clean_directory}/image_summary.csv'
img_summary_data = pd.read_csv(img_summary_path)
```

```python
#load json files
with open(demo_st_file_path, 'r') as f_st, \
     open(demo_ot_file_path, 'r') as f_ot:
    
    scene_tag_imp = json.load(f_st)
    object_tag_imp = json.load(f_ot)
```

### Find Demo Example

---

```python
#count object tag non-zero importance occurances
ot_occ = Counter()

for idx in object_tag_imp.keys():
    img_tags = object_tag_imp[idx]
    if len(img_tags) > 0:
        for t in img_tags.keys():
            ot_occ[t] += 1
```

```python
#top 10 most common non-zeros importance tags
ot_occ.most_common(20)
```

### Demo

---

```python
#target tags
tt = 'pope francis'

#output image tag, caption and thumbnail for images about pope francis
for idx in object_tag_imp.keys():
    img_tags = object_tag_imp[idx]
    if tt in img_tags.keys():
        print(f'Image ID: {idx} \n')
        #outout image caption
        print('Caption:')
        print(img_summary_data[img_summary_data.id == idx].summary.values, '\n')
        print('Object Tag Importance: ')
        #sort scene tag importance
        ots = sorted(img_tags, key=img_tags.get, reverse=True)
        for ot in ots:
            print(f'{ot}: {img_tags[ot]:.3f}')
        print()
        print('Scene Tag Importance: ')
        #sort scene tag importance
        sts = sorted(scene_tag_imp[idx], key=scene_tag_imp[idx].get, reverse=True)
        for st in sts:
            print(f'{st}: {scene_tag_imp[idx][st]:.3f}')
        #output thumbnail if possible
        try:
            img_path = f'{thumbnail_directory}/{idx}.jpg'
            fig, ax = plt.subplots(figsize = (6,6))
            img = plt.imread(img_path)
            ax.imshow(img)
            plt.show()
        except:
            print('\nThumbnail unavailable')
        print()
```
